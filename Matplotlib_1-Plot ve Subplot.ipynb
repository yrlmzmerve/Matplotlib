# Matplotlib _ Plot ve Subplot

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x=["1.ABD","2.ÇİN","3.JAPONYA","4.ALMANYA","5.İNGİLTERE","6.FRANSA"]
y=[20410,14090,5170,4210,2940,2930]
z=[63039,9957,40649,51158,44162,44915]
```

```python
plt.plot(x,z)
```

```python
# Grafiği renklendirmek için : "g" yeşil, "r" kırmızı
# "p" : nokta , "y" sarı
plt.plot(x,y,"y")
plt.xlabel("Ülkeler")
plt.ylabel("Milli Gelir")
plt.title("Ülkelere Göre Milli Gelir")
plt.show()
```

# SUBPLOT _ iç içe iki grafik yapmak için

```python
# subplot(1,2,1) : 1 satırlı 2 sutunlu grafik olustur bunun 1.isi için sunu sunu yap 
plt.subplot(1,2,1)
plt.plot(x,y,"b--")
plt.subplot(1,2,2) # 1 satır 2 sutununlunun 2.cisi
plt.plot(x,z,"r*-")
plt.tight_layout() # grafiklerin iç içe girmesini önler
```

# Figure

```python
f=plt.figure()
#([0.1,0.1,0.9,0.9]) : oluşturulan tablonun boyutları için
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel("Ülkeler")
axes.set_ylabel("Milli Gelir")
axes.set_title("Ülkelere Göre Milli Gelir")
```

## İç içe iki grafiği koyma

```python
f=plt.figure()
axes1=f.add_axes([0.1,0.1,0.9,0.9])
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,"y*-")
axes2.plot(x,z,"r")
```

```python
#grafik oluşturma baska bir yöntem
# nrow-ncols = kaç satır kaç sutun olsun bilgisi

fig,axes=plt.subplots(nrows=1, ncols=2)


axes[0].plot(x,y,"b-*")
axes[0].set_title("Milli Gelir")

axes[1].plot(x,z,"y")
axes[1].set_title("Kişi Başına Gelir")

plt.tight_layout()
```

### Grafiğe özellikler ekleme
#### renk , çizgi boyutu

```python
# linewidth : kalınlık 
# linestyles = : , -- , - , -. 
# marker = o , * , +
#alpha = saydamlaştırma

f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="purple",linewidth=1,linestyle="--", 
          marker="*", markersize=20, markerfacecolor="b", 
          markeredgewidth=2, markeredgecolor="g", alpha=0.8)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
# Ülkelerde 1'den 3 e kadar almak için : axes.set_xlim([0,2])

f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red",linewidth=1,linestyle="--", marker="*")
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")

axes.set_xlim([2,6])
axes.set_ylim([2500,5500])
```

### Grafik boyutu ve kaydetme

```python
# figsize : boyut w-h

f=plt.figure(figsize=(10,6))
axes=f.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("Ülkeler")
axes.set_ylabel("Milli Gelir")
axes.set_title("Ülkelere Göre Milli Gelir")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2,figsize=(13,5))

axes[0].plot(x,y,"b")
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x,z,"y")
axes[1].set_title("KİŞİ BAŞINA")

plt.tight_layout()
```

## 1 grafikte 2 farklı veri tipini görme
#### 2 grafiği iç içe ekleme

```python
#legend : hangi çizginin hangisine ait oldugu belirtir grafikte

fig=plt.figure(figsize=(10,5))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,"r--",label="milli gelir")
ax.plot(x,z,"y",label="kişi başına")

ax.legend(loc=5)
```

# piecharts

```python
labels = 'TÜRKİYE', 'ABD', 'ÇİN', 'ALMANYA'
sizes = [15, 35, 35, 15]
explode = (0.2, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
